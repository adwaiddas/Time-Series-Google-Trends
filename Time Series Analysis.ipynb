```python
## Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
## Import Google Data
df=pd.read_csv('C:/Users/91782/Documents/Python Scripts/multiTimeline.csv',skiprows=1)
df.head()
```

```python
## Rename the Columns
df.columns = ['Date','Covid19','Hospital','Vaccine']
df.head()
```

```python
## Find the data types
df.dtypes
```

```python
## Convert Date datatype to DateTime
from datetime import datetime

df['Date']=pd.to_datetime(df.Date)
```

```python
df.head()
```

```python
## Replace values <1 with 0
df['Covid19']=df['Covid19'].replace('<1',0)

df['Vaccine']=df['Vaccine'].replace('<1',0)

df.head()
```

```python
df.dtypes
```

```python
## Convert datatype to int
df['Covid19']=df['Covid19'].astype(int)
df['Vaccine']=df['Vaccine'].astype(int)
```

```python
df.dtypes
```

```python
## Plot Rolling Average of Covid19 search
covid=df['Covid19']

covid.rolling(12).mean().plot(figsize=(20,10),linewidth=2,fontsize=20,color='blue')

plt.xlabel('Year',fontsize=20)
```

```python
## Plot Rolling Average of Vaccine search
vaccine=df['Vaccine']

vaccine.rolling(12).mean().plot(figsize=(20,10),linewidth=2,fontsize=20,color='red')

plt.xlabel('Year',fontsize=20)
```

```python
## Plot Rolling Average of Hospital search
hospital=df['Hospital']

hospital.rolling(12).mean().plot(figsize=(20,10),linewidth=2,fontsize=20,color='yellow')

plt.xlabel('Year',fontsize=20)
```

```python
## Plot all searches
df.plot(figsize=(20,10),linewidth=2,fontsize=20)

plt.xlabel('Year',fontsize=20)
```

```python
## Find out Correlation of columns
df.corr()
```

```python
## Plot for hospital search
hospital.diff().plot(figsize=(20,10),linewidth=2,fontsize=20)

plt.xlabel('Year',fontsize=20)
```

```python
## Plot for all search
df.diff().plot(figsize=(20,10),linewidth=2,fontsize=20)

plt.xlabel('Year',fontsize=20)
```
